```python
import os
import duckdb

duckdb_path = os.getenv("DUCKDB_PATH")
assert duckdb_path is not None, "Variable 'DUCKDB_PATH' must be defined"

con = duckdb.connect(duckdb_path)
con.sql("SELECT * FROM INFORMATION_SCHEMA.TABLES;").df()
```

```python
most_voted_themes_query = open("../queries/most_voted_themes.sql").read()

most_voted_themes = con.execute(most_voted_themes_query).df()

most_voted_themes
```

```python
deputy_count_per_theme_query = open("../queries/deputy_count_per_theme.sql").read()

deputy_count_per_theme = con.execute(deputy_count_per_theme_query).df()

deputy_count_per_theme
```
